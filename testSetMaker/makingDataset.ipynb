```python
def adjust_gamma(image, gamma=1.0):
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)])
    return cv2.LUT(image.astype(np.uint8), table.astype(np.uint8))
```

```python
def predYOLO(image):
    yolo = YoloDetector(target_size=720, device="cuda:0", min_face=90)
    
    bboxes,_ = yolo.predict(image)
    
    
        
    if len(bboxes[0]) == 1:
        
        try:
            
            startX, startY, endX, endY = bboxes[0][0][0], bboxes[0][0][1],bboxes[0][0][2],bboxes[0][0][3]
            return startX, startY, endX, endY
        
        except Exception as e:
            return -1,-1,-1,-1
    
    
    else :
        return -1,-1,-1,-1       
            

```

```python
files = os.listdir(image_directory)

df = pd.DataFrame(columns = ['name','x1', 'x2','y1','y2','classname'])

for f in files:
    splitted = f.split('|')
    image =  cv2.imread(os.path.join(image_directory,f),1)
    
    

    if splitted[0] != 'face_with_mask_incorrect':
        image = cv2.resize(image, (992,720))
        startX, startY, endX, endY = predYOLO(image)
        if startX != -1:
            df = df.append({'name' : f + '.jpg', 'x1' : startX, 'x2' : endX, 'y1':startX, 'y2':endX, 'classname':splitted[0]},ignore_index = True)
    else :
        df = df.append({'name' : f + '.jpg', 'x1' : 0, 'x2' : 0, 'y1':0, 'y2':0, 'classname':splitted[0]},ignore_index = True)
    
    cv2.imwrite(output + f + '.jpg', image)

    df.to_csv('test.csv')

```
